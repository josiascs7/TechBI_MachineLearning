Las estadísticas descriptivas incluyen aquellas que resumen la tendencia central, la dispersión y la forma de la distribución de un conjunto de datos, excluyendo los valores de NaN.

Antes de generar estadísticas sobre nuestro conjunto de datos, si observa la columna 'Categoría de edad', notará que los valores únicos están en el rango de [a-b] donde a es el mínimo y b es el máximo.
- Categoría de edad: ['18-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59 ', '60-64', '65-69', '70-74', '75-79', '80 o mayor']

Tenemos que codificar la columna y hacer algunos cálculos para que pueda usarse para generar estadísticas.
- Uno de los métodos más fáciles para convertir [a-b] en un número entero es calcular la media y reemplazarla con el valor original.
- Por ejemplo, si tenemos [55-59], la media es 57. Por lo tanto, todos los valores del 55 al 59 se reemplazarán por 57.

```python
# encode 'AgeCategory' column
def media_rango(rango):
    if rango == '80 or older':
        return 80
    else:
        a, b = rango.split('-')
        return int( (int(a)+int(b)) / 2 )

#data['AgeCategory']=data['AgeCategory'].apply(lambda x: 80 if x=='80 or older' else int((int(x.split('-')[0]) + int(x.split('-')[1])) / 2))

data['AgeCategory']=data['AgeCategory'].apply(media_rango)
data['AgeCategory']
```

### 5.4 Handling Imbalanced Data
Conjunto de datos equilibrado vs desequilibrado:

- Conjunto de datos equilibrado: en un conjunto de datos equilibrado, hay una distribución aproximadamente igual de clases en la columna de destino.
- Conjunto de datos desequilibrado: en un conjunto de datos desequilibrado, hay una distribución de clases muy desigual en la columna de destino.

Echemos un vistazo a nuestro objetivo, que es 'HeartDisease':

```python
df['HeartDisease'].value_counts()
```

Como mostramos en el código anterior, hay 292k 'No' y 27k 'Sí', entonces representa un conjunto de datos desequilibrado ya que hay una distribución muy desigual de las dos clases.
<br>
Problema con conjunto de datos desequilibrado:

- Los algoritmos pueden estar sesgados hacia la clase mayoritaria y, por lo tanto, tienden a predecir la salida como la clase mayoritaria.
- Las observaciones de la clase minoritaria parecen ruido para el modelo y el modelo las ignora.
- El conjunto de datos desequilibrado proporciona una puntuación de precisión engañosa.
<br>
Técnicas para tratar con conjuntos de datos desequilibrados:

- Bajo muestreo
   - En esta técnica, reducimos el tamaño de la muestra de la clase Mayoritaria y tratamos de igualarlo con el tamaño de la muestra de la Clase Minoritaria.
     - Por ejemplo, tome un conjunto de datos de entrenamiento desequilibrado con 1000 registros.
     - Antes de submuestreo:
       - Clase de destino 'Sí' = 900
       - Clase de destino 'No' = 100
     - Después de un muestreo insuficiente:
       - Clase de destino 'Sí' = 100
       - Clase de destino 'No' = 100
     - Ahora, ambas clases tienen el mismo tamaño de muestra.

- Sobre muestreo
   - En esta técnica, aumentamos el tamaño de la muestra de la clase Minoritaria por replicación e intentamos igualarlo con el tamaño de la muestra de la Clase Mayoritaria.
     - Por ejemplo, tomemos el mismo conjunto de datos de entrenamiento desequilibrado con 1000 registros.
     - Antes de submuestreo:
       - Clase de destino 'Sí' = 900
       - Clase de destino 'No' = 100
     - Después de un muestreo insuficiente:
       - Clase de destino 'Sí' = 900
       - Clase de destino 'No' = 900
     - Ahora, ambas clases tienen el mismo tamaño de muestra.

Usaremos ambos métodos en nuestro proyecto, así que usamos 'SMOTE' para over_sampling y 'NearMiss' para under_sampling.

```python
# import SMOTE from imblearn.over_sampling
from imblearn.over_sampling import SMOTE
```

```python
from imblearn.under_sampling import NearMiss 
```

### 5.5 K-Fold Cross Validation
K-Fold CV es donde un conjunto de datos dado se divide en un número K de secciones/pliegues donde cada pliegue se usa como un conjunto de prueba en algún momento.
<br>
Tomemos el escenario de la validación cruzada de 5 veces (K = 5). Aquí, el conjunto de datos se divide en 5 pliegues. En la primera iteración, el primer pliegue se usa para probar el modelo y el resto se usa para entrenar el modelo. En la segunda iteración, el segundo pliegue se usa como conjunto de prueba, mientras que el resto sirve como conjunto de entrenamiento. Este proceso se repite hasta que cada pliegue de los 5 pliegues se haya utilizado como conjunto de prueba.

```python
# kfold cross validation
from sklearn.model_selection import KFold, cross_val_score, cross_val_predict

# make a 10 fold cross validation

cv = KFold(n_splits = 10, shuffle=True, random_state=1)
cv
```

# 6. Model Training

En esta sección, comparamos la fuerza de clasificación de AdaBoost, Random Forest, Decision Tree, Naïve Bayes, Regresion Logistica y SVM. Después de entrenar nuestros modelos, Naïve Bayes logró la precisión más alta, mientras que Perceptron alcanzó la precisión más baja.


```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sn

def train_model(model_name, params, X_train, y_train, X_test, y_test, ax_cm=None, ax_roc=None):
    if model_name == 'Decision Tree':
        model = DecisionTreeClassifier(**params)
    elif model_name == 'AdaBoost':
        model = AdaBoostClassifier(**params)
    elif model_name == 'Random Forest':
        model = RandomForestClassifier(**params)
    elif model_name == 'Naïve Bayes':
        model = GaussianNB(**params)
    elif model_name == 'KNN':
        model = KNeighborsClassifier(**params)
    elif model_name == 'Perceptron':
        model = Perceptron(**params)
    else:
        raise ValueError("Invalid model name. Please choose from Decision Tree, AdaBoost, Random Forest, Naïve Bayes, KNN, or Perceptron.")
    
    # Train the model
    
    model.fit(X_train,y_train)
    
    # Make predictions
    y_pred = model.predict(X = X_test)

    # Evaluate the model

    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    
    if ax_cm:
        sn.heatmap(cm, annot=True, linewidths=0.8, fmt='d', cmap='Blues', ax=ax_cm, cbar=False)
        ax_cm.set_xlabel('Predicted', fontsize=16)
        ax_cm.set_ylabel('Truth', fontsize=16)
        ax_cm.set_title(f"{model_name} (Acc: {accuracy:.2f})", fontsize=18, pad=20)
    
    # Plot ROC curve
    if ax_roc and hasattr(model, 'predict_proba'):
        y_score = model.predict_proba(X_test)
        yes_probs = y_score[:, 1]
        fpr, tpr, _ = roc_curve(y_test, yes_probs)
        roc_auc = auc(fpr, tpr)
        ax_roc.plot(fpr, tpr, marker='_', label=f'{model_name} (AUC = {roc_auc:.2f})')
        ax_roc.set_title(f"ROC Curve - {model_name}", fontsize=18, pad=20)

    
    print(f"Classification report <{model_name}>")
    print(report)
    
    return model
```

```python
models = [
    {'name': 'Decision Tree', 'params': {'criterion': 'entropy', 'max_depth': 100, 'random_state': 1024}},
    {'name': 'AdaBoost', 'params': {'n_estimators': 50, 'learning_rate': 1}},
    {'name': 'Random Forest', 'params': {'criterion': 'entropy'}},
    {'name': 'Naïve Bayes', 'params': {'var_smoothing': 1e-9}},
    {'name': 'KNN', 'params': {'n_neighbors': 2}},
    {'name': 'Perceptron', 'params': {'random_state': 2, 'tol': 1e-4}},
]

# Plotting confusion matrices and ROC curves for all models in a 3x3 grid
fig, axes = plt.subplots(3, 4, figsize=(20, 15))
cm_axes = axes[:, :2].ravel()
roc_axes = axes[:, 2:].ravel()

# Plot "No-Skill" on ROC Curve
for ax in roc_axes:
    ax.plot([0, 1], [0, 1], linestyle='--', label='No Skill')
    ax.set_xlabel("False Positive Rate", fontsize=14)
    ax.set_ylabel("True Positive Rate", fontsize=14)

for idx, model_info in enumerate(models):
    train_model(model_info['name'], model_info['params'], X_train, y_train, X_test, y_test, ax_cm=cm_axes[idx], ax_roc=roc_axes[idx])

# Set the legend for ROC curves
roc_axes[0].legend()

fig.tight_layout()
plt.subplots_adjust(wspace=0.4, hspace=0.6)  # Add margins between subplots
plt.show()
```
